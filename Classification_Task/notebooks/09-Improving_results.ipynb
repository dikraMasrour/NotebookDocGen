```python
def remove_ntb(classe1,classe2,list_ntb,data) :
    df1 = list_ntb[list_ntb['category'] == classe1] 
    List_to_delete_1 = []
    for i in df1.index:
        if classe1 in df1.loc[i]['title'] and classe2 in df1.loc[i]['title']:
            List_to_delete_1.append(df1.loc[i]['title'])

    df2 = list_ntb[list_ntb['category'] == 'classe2'] 
    List_to_delete_2 = []
    for i in df2.index:
        if classe1 in df2.loc[i]['title'] and classe2 in df2.loc[i]['title']:
            List_to_delete_2.append(df2.loc[i]['title'])

    for i in list_ntb.index:
        if list_ntb.loc[i]['title'] in List_to_delete_1 or list_ntb.loc[i]['title'] in List_to_delete_2 :
            list_ntb.drop(index = i,inplace=True)

    final_list = []
    for i in list_ntb.index:
        final_list.append(re.sub("^[^/]*/",'',list_ntb.loc[i]['title'])+".ipynb")


    count_1 = 0
    count_2 = 0

    for i in data.index:
        if data.loc[i]['title'] in final_list:
            count_1 +=1
        else :
            data.drop(index = i,inplace=True)
            count_2 +=1

    print("Number of Removed cells :",count_2)
    print("Removed notebooks \n ",List_to_delete_1 +List_to_delete_2)
```

```python
new_df = pd.DataFrame(columns = ['markdown','code','title'])
```

```python
List_ntb = []
for i in data.index:
    List_ntb.append(data.loc[i]['title'])

List_ntb_only = np.unique(List_ntb)
```

```python
def grp_pairs(data,title,limit,new_df):
    i = 0
    
    if i < limit-1 and data.loc[i]['cell-type'] == 'markdown' and data.loc[i+1]['cell-type'] == 'code':
        new_df.loc[len(new_df)] = [data.loc[i]['source'],data.loc[i+1]['source'],title]
        i+=2
        if i < limit :
            grp_pairs(data[i:].reset_index(drop=True),title,len(data[i:]),new_df)

    else :
            i+=1
            if i < limit :
                grp_pairs(data[i:].reset_index(drop=True),title,len(data[i:]),new_df)
```

```python
for ntb in List_ntb_only :
    data_ = data[data['title'] == ntb].reset_index(drop=True)

    try:
        grp_pairs(data_,ntb,len(data),new_df)
    
    except Exception as e:
        print(ntb)
```
