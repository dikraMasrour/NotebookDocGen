Importing libs

```python
import pandas as pd
import nltk
import pickle
import re
import string
from nltk.tokenize import RegexpTokenizer
from langdetect import detect, DetectorFactory
import pyparsing
```

Path to data

```python
DATA_PATH ='../data/first_data_classification.csv'
DetectorFactory.seed = 0
PATTERN = r'[' + string.punctuation +'\n' ']'
COMMENT_FILTER = pyparsing.pythonStyleComment.suppress()
```

```python
raw_df = pd.read_csv(DATA_PATH).drop(['Unnamed: 0'],axis=1)
raw_df
```

- Remove punctuations and special caracters from md and non-en noteboks
- Remove comments from code

```python
df = pd.DataFrame(columns=['all_md', 'all_cd', 'tag'])
```

```python
for i in raw_df.index:
    raw_df.loc[i]['all_md'] = re.sub(PATTERN,'',str(raw_df.loc[i]['all_md']))
    try:
        if detect(raw_df.loc[i]['all_md']) == 'en':
            raw_df.loc[i]['all_cd'] = COMMENT_FILTER .transformString(raw_df.loc[i]['all_cd'])
            df.loc[len(df)]=raw_df.loc[i]

    except Exception as e:
            print('exception')
```
