Importing libs

```python
import pandas as pd
import nltk
import pickle
import re
import string
import pyparsing
import warnings
import mdtex2html
from langdetect import detect, DetectorFactory
from bs4 import BeautifulSoup
from cleantext import clean
```

Path to data

```python
DATA_PATH ='../data/first_data_classification.csv'
PATTERN = r'[' + string.punctuation +'\n' ']'
COMMENT_FILTER = pyparsing.pythonStyleComment.suppress()
RMV_URL = re.compile(r'https?://\S+|www\.\S+')
DetectorFactory.seed = 0
warnings.filterwarnings('ignore')
```

```python
raw_df = pd.read_csv(DATA_PATH).drop(['Unnamed: 0'],axis=1)
raw_df
```

```python
raw_df.isna().sum()
```

```python
for i in raw_df.columns:
    raw_df[i].fillna(' ', inplace=True)
```

```python
df = pd.DataFrame(columns=['all_md', 'all_cd', 'tag'])
```

- Convert latex caracters to html
- Remove html caracters
- Remove punctuations and special caracters from md and non-en noteboks
- Remove comments from code

```python
for i in raw_df.index:
    
    raw_df.loc[i]['all_md'] = mdtex2html.convert(str(raw_df.loc[i]['all_md']))
    raw_df.loc[i]['all_md'] = BeautifulSoup(str(raw_df.loc[i]['all_md']), "lxml").text
    raw_df.loc[i]['all_md'] = re.sub(PATTERN,'',RMV_URL.sub(r'',str(raw_df.loc[i]['all_md'])))
    raw_df.loc[i]['all_md'] = clean(str(raw_df.loc[i]['all_md']), no_emoji=True)
    
    if raw_df.loc[i]['all_md'] == '' :
            raw_df.loc[i]['all_cd'] = COMMENT_FILTER .transformString(raw_df.loc[i]['all_cd'])
            df.loc[len(df)]=raw_df.loc[i]
    else : 
        try:
            if detect(str(raw_df.loc[i]['all_md']).strip()) == 'en' :
                raw_df.loc[i]['all_cd'] = COMMENT_FILTER .transformString(raw_df.loc[i]['all_cd'])
                df.loc[len(df)]=raw_df.loc[i]

        except Exception as e:
                print(e)
```
